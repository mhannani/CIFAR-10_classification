# CIFAR-10 images classification using Deep Learning

The CIFAR-10 dataset consists of 60000 32x32 colour images in 10 classes, with 6000 images per class. There are 50000 training images and 10000 test images.

![cifar](Images/cifar_10.png)

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Load the data

```python
from loaders import load_cifar_10
```

```python
(x_train, y_train), (x_test, y_test) = load_cifar_10()
```

## Visualize data

```python
from visualize import display_some_images
```

```python
display_some_images(x_train, y_train, 6)
```

```python
from models import CifarModel
```

```python
model = CifarModel((32, 32, 3),[32, 32, 64, 64], [3, 3, 3, 3], [1, 2, 1, 2], 10)
```

```python
model.save('cifar_model')
```

```python
model.compile(lr=0.0005)
```

```python
model.train(x_train, y_train)
```

## Evaluate the model

```python
model.evaluate(x_test, y_test)
```

## Make prediction

```python
a = model.plot_predictions(x_test, y_test, nb_imgs=10)
```
